```python
import os
import time
import pandas as pd

from pyvrp import read, solve, stop

# ==========================================================
# Configuration
# ==========================================================

TIME_LIMIT = 30
SEEDS = [0]

# ==========================================================
# Find all X-set instances in /
# ==========================================================

vrp_files = vrp_files = [
    "X-n101-k25.vrp",
    "X-n176-k26.vrp",
    "X-n251-k28.vrp",
    "X-n491-k59.vrp",
    "X-n685-k75.vrp",
]

print(f"Found {len(vrp_files)} X instances")

# ==========================================================
# Benchmark
# ==========================================================

records = []

for vrp_file in vrp_files:

    name = vrp_file[:-4]

    print(f"\nSolving {name}")

    try:

        instance = read("/" + vrp_file)

        for seed in SEEDS:

            start = time.perf_counter()

            result = solve(
                instance,
                stop=stop.MaxRuntime(TIME_LIMIT),
                seed=seed,
                display=False,
            )

            runtime = time.perf_counter() - start

            records.append(
                {
                    "instance": name,
                    "seed": seed,
                    "cost": result.cost(),      # may need adjustment
                    "runtime_sec": runtime,
                }
            )

    except Exception as e:

        print(f"FAILED: {name}")
        print(repr(e))

df = pd.DataFrame(records)

print("\nResults:")
print(df)

df.to_csv("pyvrp_results.csv", index=False)

print("\nSaved pyvrp_results.csv")
```
